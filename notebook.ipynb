- Constraint programming based branch and price and cut algo for integrated operating room planning and scheduling (IORPS)


Assumptions:
1. Durations of surgeries are deterministic
2. Surgeon has max surgery time for the day
3. Surgeries assigned to surgeon 
4. Recovery rooms are not bottleneck
5. 8 hours max in OR, no overtime
6. Operating rooms are identical in time and equipment

Maximize:
1. Mandatory surgeries scheduled
2. Surgeons don't exceed max daily surgery time
3. OR can only handle one patient at a time

Variables:

D : Set of days in planning horizon
I_set : Set of all surgeries 
I_1 : Set of all mandatory surgeries
I_2 : Set of all optional surgeries
I_l : Set of all surgeries for surgeon l
K_d : Set of OR for day d
L : Set of surgeons
T_d Set of time slots for day d discretized

D = {{K_1, T_1}, {K_2, T_2}, ... {K_d, T_d}}

Code Subproblem (Section 3-2)
Code Master Problem (Section 3-1)
